$$
\frac {AR}{K_P} \ = \frac{1}{\sqrt{a^2 + b^2}} \  \ where  \\  a = 1 - \tau^2\omega^2 \ ; \ b = -2\tau\zeta\omega
$$
$$
\tan(\phi) \ = \ (\frac{-b}{a})
$$

```python
tau = 1.42
import numpy as np
import math
z = [0.3,1.0,2.5]
w = np.linspace(0.001,1000,10000)
a = 1-(tau**2)*(w**2)
b1 = -2*tau*w*0.3
ar1 = 1/(a**2 + b1**2)**0.5
b2 = -2*tau*w*1.0
ar2 = 1/(a**2 + b2**2)**0.5
b3 = -2*tau*w*2.5
ar3 = 1/(a**2 + b3**2)**0.5
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
%config InlineBackend.figure_format = 'svg'
style.use('classic')
plt.plot(w,ar1,'r',label = '$ \zeta = 0.3 $')
plt.plot(w,ar2,'g',label = '$ \zeta = 1.0 $')
plt.plot(w,ar3,'b',label = '$ \zeta = 2.5 $')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.xlabel(r'$ log_{10}{\omega} $',fontsize = 14)
plt.ylabel(r'$ log_{10}\frac{AR}{K_P} $',fontsize = 14)
plt.title(r'$  \frac{AR}{K_P} \ vs \ \omega \ \ for \ \ second \ \ order \ \ process $',fontsize = 18)
plt.savefig('ar.pdf')
```

```python

phi1 = np.arctan(-b1/a) 
phi2 = np.arctan(-b2/a)
phi3 = np.arctan(-b3/a)
plt.plot(w,phi1,'r',label = '$ \zeta = 0.3 $')
plt.plot(w,phi2,'g',label = '$ \zeta = 1.0 $')
plt.plot(w,phi3,'b',label = '$ \zeta = 2.5 $')
plt.xscale('log')

plt.legend()
plt.xlabel(r'$ log_{10}{\omega} $',fontsize = 16)
plt.ylabel(r'$ \phi $',fontsize = 16)
plt.title(r'$  \phi \ vs \ \omega \ \ for \ \ second \ \ order \ \ process $',fontsize = 20)
```

```python
w1 = np.linspace(0.001,1/tau,1000)
w2 = np.linspace(1/tau,1000,10000)
a1 = 1-(tau**2)*w1**2
a2 = 1-(tau**2)*w2**2
b11 = -2*0.3*tau*w1
b12 = -2*0.3*tau*w2
phi11 = np.arctan(b11/a1)
phi12 = np.arctan(-b12/a2)
b21 = -2*1.0*tau*w1
b22 = -2*1.0*tau*w2
phi21 = np.arctan(b21/a1)
phi22 = np.arctan(-b22/a2)
b31 = -2*2.5*tau*w1
b32 = -2*2.5*tau*w2
phi31 = np.arctan(b31/a1)
phi32 = np.arctan(-b32/a2)
w = np.concatenate((w1,w2))
phi1 = np.concatenate((phi11,phi12))
phi2 = np.concatenate((phi21,phi22))
phi3 = np.concatenate((phi31,phi32))
plt.plot(w,phi1,'r',label = '$ \zeta = 0.3 $')
plt.plot(w,phi2,'g',label = '$ \zeta = 1.0 $')
plt.plot(w,phi3,'b',label = '$ \zeta = 2.5 $')
plt.xscale('log')
plt.legend()
plt.xlabel(r'$ log_{10}{\omega} $',fontsize = 14)
plt.ylabel(r'$ \phi $',fontsize = 14)
plt.title(r'$  \phi \ vs \ \omega \ \ for \ \ second \ \ order \ \ process $',fontsize = 18)
plt.savefig('phi.pdf')
```
